```python
query = '''
SELECT source.identifier, source.name, target.identifier, target.name, pc.metapath_id, pc.dwpc, pc.p_value
FROM dj_hetmech_app_pathcount pc
JOIN dj_hetmech_app_node source ON pc.source_id = source.id
JOIN dj_hetmech_app_metanode source_metanode ON source.metanode_id = source_metanode.identifier
JOIN dj_hetmech_app_node target on pc.target_id = target.id
JOIN dj_hetmech_app_metanode target_metanode ON target.metanode_id = target_metanode.identifier
WHERE source_metanode.abbreviation = 'C' AND target_metanode.abbreviation = 'D'
'''

connection = psycopg2.connect(host = 'hetmech-db-dev.cobepk65dd7j.us-east-1.rds.amazonaws.com', 
                              database = 'dj_hetmech', user = 'read_only_user', password = db_password)

cursor = connection.cursor()

cursor.execute(query)
top_metapaths = cursor.fetchall()

top_metapaths = pandas.DataFrame(top_metapaths, columns=[
    'source_id', 'source_name', 'target_id', 'target_name', 'metapath', 'dwpc', 'p_value'])
```
