```python
import pandas as pd
import sklearn.metrics
```

```python
plot_df = pd.read_csv('plot_info.tsv', sep='\t')

plot_df.head()
```

```python
auroc_df = (
    plot_df
    .assign(
        feature_group = lambda df: df['feature_group'].astype(str)
    )
    .groupby(['feature', 'feature_information', 'feature_group', 'data_source', 'task'])
    .apply(lambda group: sklearn.metrics.roc_auc_score(group['edge_test'], group['value']))
    .reset_index()
    .rename(columns={0: 'AUROC'})
)

auroc_df.to_csv('auroc.csv', index=False)

auroc_df.head()
```
